# Removing SST Errors

Investigations of ORAS5 led to a `SST product` where the early-period `warm-SST bias was removed` (ORAS5m; Tietsche et al. 2020).

The existence of ORAS5m means an `experiment investigating the causal link` between SEAS5 SST and jet biases is possible:

<div class="alert alert-block alert-info">
    <b>Causality Experiment (SEAS5m):</b> Re-run SEAS5 hindcasts with improved SST initial conditions and check if jet biases are improved.
</div>

## Summary and Future Steps

Results from the causality experiment `appear to support the hypothesis` that SST biases in the NASG region in ORAS5 are `partially responsible for the jet biases` found in SEAS5 in late winter.

The next step is to try and `quantify the strength of the causal link` between SST biases and jet biases.

One approach for doing this is to `build a Granger Causal Model` of the coupling between SSTs and the jet (Mosedale et al 2006).
