# Motivation

#### 1. SEAS5 North Atlantic Subpolar Gyre `SST errors` in the early reforecast period (1981 - 1996)

- SEAS5 winter reforecasts have a warm SST bias in the North Atlantic Subpolar Gyre (NASG) region at the start of the reforecast period (1981 - 1996).
- The bias is `non-stationary`: it disappears from the 2000s onwards. See the image below from Tietsche et al. (2020).
- Sensitivity experiments show the problem lies with the ocean initial conditions provided by ORAS5.

```python
Image(filename = '../images/tietsche_2020.png', width = 900) 
```

<br>

#### 2. An `ocean-jet stream connection` in late winter

- SST variability in the NASG region has been linked (via `correlation analysis`) to jet stream variability in late winter (`late February/March`).
- Long-term variations in the NASG region have been associated with long-term variations in the jet. See the image below from Simpson et al. (2018).
